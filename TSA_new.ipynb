```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

pred = []
X = trainm.confirmed.values

for i in testm.confirmed:
    model_fit = SARIMAX(X, order=(1,0,1)).fit()
    predf = model_fit.forecast()
    pred.append(predf)
    X = np.append(X, i)

testm['Predictions'] = pred
testm
```

```python
print(model_fit.summary())
```

```python
testm['Predictions'] = testm['Predictions'].apply(np.sum)
testm
```

```python
plt.figure(num=None,figsize=(8,6),dpi=80)

plt.plot(testm.confirmed,label="Test data")
plt.plot(testm.Predictions,label="Predicted data")
plt.legend()
plt.show()
```
